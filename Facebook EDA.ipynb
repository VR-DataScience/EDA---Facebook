# <a name = Section8014></a>
## 8.1.4 How many Facebook users don’t have friends or zero friends ?

```python
# Identify facebook users with zero friends across gender
zero_friends = pd.DataFrame(fb[fb['friend_count']==0].groupby('gender')['friend_count'].count()).rename(columns={'friend_count':'zero_friends'})
zero_friends
```

```python
have_friends = pd.DataFrame(fb[fb['friend_count']>0].groupby('gender')['friend_count'].count()).rename(columns={'friend_count':'have_friends'})
have_friends
```

```python
# Total facebook users across gender
total_users = pd.DataFrame(fb['gender'].value_counts()).rename(columns={'gender':'total_users'})
total_users
```

```python
# Zero friends % across gender
no_friends = pd.DataFrame(round(zero_friends['zero_friends'] / total_users['total_users'] * 100,2)).rename(columns={0:'no_friends%'})
no_friends
```

```python
# Concate results of findings on Total facebook users, Have Friends, Zero Friends and Zero Friends %
friend_summary = pd.concat([total_users,have_friends, zero_friends,no_friends],axis=1)
friend_summary.columns=['Total Users', 'Have Friends','Zero Friends', 'Zero Friends%']
friend_summary
```

```python
# Using Bar chart showing Total Users who have friends vs Users who dont have friends
friend_summary[['Have Friends', 'Zero Friends']].plot(kind='bar', figsize=(10, 8), 
                                                                    color=colors_All,stacked=True)

plt.xlabel('User Gender')
plt.ylabel('Friends Count\n')
plt.title('No. of Users - Have Friends and Zero Friends across gender\n',fontsize=30)
plt.legend(labels=('Have Friends', 'Zero Friends \n(male-2.48%, female-1.25%)'),fontsize=14)
plt.xticks(rotation=0)
# In order to save your plot into an image on your system, use the following command.
plt.savefig("Stacked Bar chart - Zero Friends.jpeg")             # The image will be saved in the directory of this notebook.

```

**Observations:**
- Majority of facebook users have friends
- Only **2.48%** of males dont have friends and **1.25%** of females dont have friends.
